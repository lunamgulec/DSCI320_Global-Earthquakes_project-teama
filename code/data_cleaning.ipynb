```python
missing = (
    df.isnull()
    .melt(var_name='column', value_name='is_missing')
    .groupby(['column', 'is_missing'])
    .size()
    .reset_index(name='count')
)

alt.Chart(missing).mark_bar().encode(
    x='column:N',
    y='count:Q',
    color='is_missing:N',
    tooltip=['column', 'is_missing', 'count']
).properties(
    title='Missing Values per Column'
)
```

```python
df.dropna(inplace=True)
```

```python
missing = (
    df.isnull()
    .melt(var_name='column', value_name='is_missing')
    .groupby(['column', 'is_missing'])
    .size()
    .reset_index(name='count')
)

alt.Chart(missing).mark_bar().encode(
    x='column:N',
    y='count:Q',
    color='is_missing:N',
    tooltip=['column', 'is_missing', 'count']
).properties(
    title='Missing Values per Column'
)
```

```python
df.to_csv("../data/processed/cleaned_data.csv", index=False)
```

```python
alt.data_transformers.enable("vegafusion")
mag_range_chart = alt.Chart(df).mark_bar().encode(
    # Use binning (histogram) to show the range and distribution
    alt.X("mag", bin=alt.Bin(maxbins=20), title="Magnitude (mag)"),
    alt.Y("count()", title="Frequency"),
    tooltip=[alt.Tooltip("mag", bin=True), "count()"]
).properties(
    title="Range and Distribution of Earthquake Magnitude"
)
mag_range_chart 
```

```python
bins = [0.0, 4.0, 5.0, 6.0, 7.0, np.inf] 

# Define the labels for the bins (in ascending order)
labels = ['Minor (<4.0)', 'Light (4.0-4.9)', 'Moderate (5.0-5.9)', 'Strong (6.0-6.9)', 'Major (>=7.0)']

# Create the Ordinal Column 
df['mag_ordinal'] = pd.cut(
    df['mag'], 
    bins=bins, 
    labels=labels, 
    right=False,  
    ordered=True
)

print("--- Magnitude Ordinal Transformation Results ---")
print("New Column Data Type:")
print(df['mag_ordinal'].dtype)
print("\nValue Counts (Showing Order):")
print(df['mag_ordinal'].value_counts(sort=False))
```

```python
alt.data_transformers.enable("vegafusion")
mag_range_chart = alt.Chart(df).mark_bar().encode(
    # Use binning (histogram) to show the range and distribution
    alt.X("depth", bin=alt.Bin(maxbins=20), title="Depth (km)"),
    alt.Y("count()", title="Frequency"),
    tooltip=[alt.Tooltip("depth", bin=True), "count()"]
).properties(
    title="Range and Distribution of Earthquake Depth"
)
mag_range_chart 
```

```python
bins = [0.0, 70.0, 300.0, np.inf]

# Define the labels for the bins (in ascending order)
labels = ['Shallow (0-70 km)', 'Intermediate (70-300 km)', 'Deep (>300 km)']

# Create the Ordinal Column 
df['depth_ordinal'] = pd.cut(
    df['depth'],
    bins=bins,
    labels=labels,
    right=False,  # Intervals are [lower, upper)
    ordered=True  # Ensure the data type is ordered categorical (ordinal)
)

print("--- Depth Ordinal Transformation Results ---")
print("New Column Data Type:")
print(df['depth_ordinal'].dtype)
print("\nValue Counts (Showing Order):")
print(df['depth_ordinal'].value_counts(sort=False))
```
